```python
##import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x=6 * np.random.rand(100,1)-3
y=0.5 * x**2 + 1.5*x + 2 + np.random.randn(100,1)
plt.scatter(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
##lets impliement simple liner regression
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
X_train
```

```python
regression_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression_1.predict(X_test))
print (score)
```

```python
##lets vizulaize this model
plt.plot(X_train,regression_1.predict(X_train),color='red')
plt.scatter(X_train,y_train,color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#lets apply polynomial transformaton
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2, include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,regression.predict(X_test_poly))
print (score)
```

```python
print (regression.coef_)
print (regression.intercept_)
```

```python
plt.scatter(X_train,regression.predict(X_train_poly),color='red')
plt.scatter(X_train,y_train,color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
poly=PolynomialFeatures(degree=2, include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
#predict new data set
X_new=np.linspace(-3,3,100).reshape(100,1)
X_new_poly=poly.transform(X_new)
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,regression.predict(X_test_poly))
print (score)
```

```python
y_new=regression.predict(X_new_poly)
plt.plot(X_new,y_new,"r-", linewidth=2,label="prediction")
plt.plot(X_train,y_train,"b.",label="training data")
plt.plot(X_test,y_test,"g.",label="test data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```
